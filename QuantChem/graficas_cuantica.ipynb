```python
import numpy as np
import matplotlib.pyplot as plt

#Función a estudiar: Armónico esférico con números cuánticos 3,3

Cnorm = -(1/8)*np.sqrt(35/np.pi)  #constante de normalización

def Y2(theta,phi): #Y cuadrado
    return abs((Cnorm**2)*np.sin(theta)**6)

def ReY(theta,phi): #parte real de Y
    return abs(Cnorm*(np.sin(theta)**3)*np.cos(3*phi))
    
def ImY(theta,phi): #parte imaginaria de Y
    return abs(Cnorm*(np.sin(theta)**3)*np.sin(3*phi))
```

A continuación se grafican diferentes cortes de la función $Y^2$, esto se consigue utilizando $\theta$ constante para cada gráfica. Estos cortes corresponden a cortes con un plano horizontal.

$\theta$ varía entre 0 y $\pi$

```python
phi = np.linspace(0,2*np.pi,200)
graphs = 5
fixed_theta = [i*np.pi/(graphs-1) for i in range(graphs)] #varía theta para obtener diferentes cortes

scale = 0.2
for i in fixed_theta:
    Y2s = [Y2(i,t) for t in phi]  #puntos a graficar para cierto theta constante
    ax = plt.subplot(projection='polar')
    ax.plot(0,scale, label = 'phi = '+str(i/np.pi)+'pi')  #arregla la escala, para que todos los cortes tengan la misma
    ax.plot(phi,Y2s)
    ax.legend()
    plt.show()
```

La función no depende de $\phi$. Por lo tanto se obtienen círculos en cada corte a theta constante. La siguiente es una gráfica en la que la variable es $\theta$ (un corte vertical). Se elige $\phi = 0$ por facilidad, pues la función no depende de $\phi$.

En la lista de datos utilizada para los valores de $Y^2$ en función de $\theta$ se divide el angulo $\theta$ entre 2 para producir un corte vertical del lóbulo,
pues $\theta$ varía entre 0 y $\pi$, no entre 0 y $2\pi$ como se tiene en la escala establecida.

Esto produce una gráfica que corresponde al corte buscado, lo que permite reconstruir más fácilmente la figura tridimensional 
correspondiente al cuadrado de la función armónica esférica.

```python
theta = np.linspace(0,2*np.pi,100)
Y2s = [Y2(t/2,0) for t in theta]  #puntos a graficar para phi = 0

scale = 0.15
ax = plt.subplot(projection='polar')
ax.plot(0,scale)  #arregla la escala, para que todos los cortes tengan la misma
ax.plot(theta,Y2s)
plt.show()
```

A continuación se grafican diferentes cortes de la función $Re(Y)$, esto se consigue utilizando $\theta$ constante para cada gráfica.

$\theta$ varía entre 0 y $\pi$

```python
phi = np.linspace(0,2*np.pi,200)
graphs = 5
fixed_theta = [i*np.pi/(graphs-1) for i in range(graphs)]

scale = 0.4
for i in fixed_theta:
    ReYs = [ReY(i,t) for t in phi]
    ax = plt.subplot(projection='polar')
    ax.plot(0,scale)
    ax.plot(phi,ReYs)
    plt.show()
```
