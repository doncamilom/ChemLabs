```python
def fitin(xvals,yvals,n,men=0,mas=0):
    coefs=list(reversed(plt.polyfit(xvals,yvals,n)))
    fitx=[i for i in plt.arange(xvals[0]-men,xvals[-1]+mas,0.001)]
    fity=[]
    for x in fitx:
        suma=0
        for e in range(len(coefs)):
            term=0
            term=coefs[e]*x**e
            suma+=term
        fity.append(suma)
    return fitx,fity,coefs[0],coefs[1]
def Tfus(a,b,c,d):
    return (a*d-b*c)/(d-b)
```

```python
def newt(val=0.1): #Algoritmo de Newton-Raphson:  P = x-f(x)/f'(x)
    for i in range(100):
        val=val-((log(val,e)/dHnaf)-(log(1-val,e)/dHdfa)-(1/Tfnaf -1/Tfdfa)/R)/((1/dHnaf +1/dHdfa)/val)
    return val
xeu=newt()

Teutteo=1/((1/(Tnaf))-(log(xeu,e)*R/dHnaf))-273.15
print('x=',1-xeu)
print('T=',round(Teutteo,2),'C')
```

A partir de la ecuacin terica dada:
\begin{equation}
ln(x)=\frac{-\Delta H}{R}\left(\frac1{T}-\frac1{T_{f,puro}}\right)    
\end{equation}

Usando $x_{naf}=1-x_{DFA}$, reemplazando en (1), despejando la variable T e igualando, se llega a lo siguiente:

\begin{equation}
\frac{ln(x_2)}{\Delta H_2}-\frac{ln(1-x_2)}{\Delta H_1}=\frac1{R}\left(\frac1{T_2}-\frac1{T_1}\right)
\end{equation}

Donde el subndice 1 se usa para Difenilamina, y el subndice 2 para Naftaleno.

El valor de $x_2$ que satisface esta ecuacin para los $\Delta H$ y $T_f$ dados, se halla con ayuda de un computador mediante el algoritmo de Newton-Raphson:

\begin{equation}
P=x-\frac{f(x)}{f'(x)}
\end{equation}

Con lo que se obtiene:

$x_{eut}=0.3601$

Despejando T de (1), se llega a:

\begin{equation}
T={\left(\frac1{T_{f,2}}-\frac{Rln(x_{eut})}{\Delta H_2}\right) }^{-1}
\end{equation}

De lo que se obtiene, reemplazando $x_2$:

$T_{eut}=31.92C$

Estas son la temperatura y concentracin eutcticas para esta mezcla.

```python
def er(teo,exp):
    return round(abs(teo-exp)*100/teo,2)
def erProm(x,T,svte): #Da el error promedio para los puntos tomados de T vs concentracion
    errores=[]
    for i in range(len(x)):
        if svte=='DFA':
            Tteo=1/((1/Tfnaf)-log(1-x[i],e)*R/dHnaf)-273.15
        else:
            Tteo=1/((1/(Tfdfa))-(log(x[i],e)*R/dHdfa))-273.15
        errores.append(er(Tteo,T[i]))
    return sum(errores)/len(errores)

e1=erProm(xnaf,Tfes[0:5],'naf')
e2=erProm(xnafi,Tnafi,'DFA')
print(e1,',',e2)
er(Teutteo,Teut)
```
