```python
from pandas import read_excel
import matplotlib.pyplot as plt
from numpy import log,poly1d
import numpy as np
from scipy import stats
```

```python
datos = read_excel('../Documents/chem/fisicoquimica/informes/visco/datos_visc.xlsx')
data = datos.as_matrix()
```

```python
visc_1 = data[3:33,2]/10
tau_1 = data[3:33,1]/10
srate_1 = tau_1/visc_1
ax1 = plt.subplot(111)
ax1.plot(srate_1,visc_1)
ax1.set_title('Curva viscosidad mayonesa a 25°C',fontsize=14)
ax1.set_xlabel(r'Shear rate $\dot{\gamma}$ $(s^{-1})$',fontsize=12)
ax1.set_ylabel(r'Viscosidad $\eta$ $(cPa.s)$',fontsize=12)
ax1.grid()
#plt.savefig('viscosidad',dpi=200)
plt.show()
```

```python
log_srate_1 = [log(i) for i in srate_1]
log_tau_1 = [log(i) for i in tau_1]

ax2 = plt.subplot(111)
ax2.plot(log_srate_1,log_tau_1,'b+')

vals = stats.linregress(log_srate_1,log_tau_1)
p= poly1d((vals[0],vals[1]))

xs = [np.amin(log_srate_1),np.amax(log_srate_1)]
ax2.plot(xs,[p(i) for i in xs],'k-',linewidth = 1, label = r'$R^2$ = ' + str(round(vals[2]**2,4)))

ax2.set_title('')
ax2.set_ylabel(r'$Ln(\tau)$')
ax2.set_xlabel(r'$Ln(\dot{\gamma})$')
ax2.grid()
plt.legend()
plt.show()

vals
```
