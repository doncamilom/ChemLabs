```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('./data.xlsx',header=None).T

muestra = data.iloc[:,0].ffill()
data.index = [muestra , data.iloc[:,2]]
data = (data.iloc[:,4:]
        .dropna())

data
```

```python
blanco = data.loc['Blanco']
Ib = blanco.iloc[1] #Corriente blanco

for m in muestra.unique():
    DataMue = data.loc[m]
    plt.plot(DataMue.iloc[0],
             DataMue.iloc[1]-Ib,
             label=m,
             color='r')
    plt.legend()
    plt.show()
```
